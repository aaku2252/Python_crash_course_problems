```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
        
def takeInput(arr) :
    head = None
    tail = None

    datas = arr

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head

def printLL(head):
    while head is not None:
        print(head.data,end=" ")
        head = head.next
    print()



def skipMdeleteN(head, m, n):
    if head is None:
        return None
    if m ==0:
        return None
    if n ==0:
        return head
    curr = head
    while curr:
        for i in range(m-1):
            if  curr.next is not None:
                curr = curr.next
            else:
                break
        tail = curr
        for j in range(n):
            if  tail.next is not None:
                tail = tail.next
            else:
                break
        curr.next = tail.next
        if curr.next is not None:
            curr = curr.next
        else:
            break
      
    return head       

node = takeInput([10, 22, 10, 26, 11, 7, 8, 1, 23, 5, 1, 28, 0,5,3,4,5 ,-1])
head = skipMdeleteN(node,2,3)
printLL(head)
```
