```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None
        
def takeInput(arr) :
    head = None
    tail = None

    datas = arr

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head

def printLL(head):
    while head is not None:
        print(head.data,end=" ")
        head = head.next
    print()

def kReverse(head,k):
    if k ==0 or k==1:
        return head
    current = head
    fwd = None
    prev = None
    
    count = 0
    
    while (count<k) and (current is not None):
        fwd = current.next
        current.next = prev
        prev = current
        current = fwd
        count+=1
    
    if fwd is not None:
        head.next = kReverse(fwd,k)
    return prev
      

node = takeInput([10, 22, 10, 26, 11, 7, 8, 1, 23, 5, 1, 28, 0,5,3,4,5 ,-1])
head = kReverse(node,2)
printLL(head)
```
