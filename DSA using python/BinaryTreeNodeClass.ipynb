```python
class BST:
    def __init__(self):
        self.root = None
        self.numNodes = 0

    def printTreeHelper(self,root):
        if root ==None:
            return
        print(root.data,end =':')
        if root.left!=None  :
            print('L',end = '')
            print(root.left.data,end = ',')
        if root.right!=None :
            print('R',end = '')
            print(root.right.data,end = ',')
        print()
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)
        
    def printTree(self):
        self.printTreeHelper(self.root) 
        
        
        

    def _search(self,root,data):
        if root ==None:
            return False
        if root.data == data:
            return True
        if root.data>data:
            return self._search(root.left,data)
        else:
            return self._search(root.right,data)
        
    def search(self, data):
        return self._search(self.root,data)
        
        
    def _insert(self, root,data):
        if root == None:
            node = BinaryTreeNode(data)
            return node
        if root.data>=data:
            root.left = self._insert(root.left,data)
            return root 
        else:
            root.right = self._insert(root.right,data)
            return root 
         
    def insert(self, data):
        self.numNodes+=1
        self.root= self._insert(self.root,data)



    def min(self, root):
        if root ==None:
            return 10000
        if root.left ==None:
            return root.data 
        return self.min(root.left)

    def _delete(self, root,data):
        if root ==None:
            return False,None
        if root.data<data:
            deleted,newRightNode = self._delete(root.right,data)
            root.right = newRightNode
            return deleted,root
        
        if root.data>data:
            deleted,newLeftNode = self._delete(root.left,data)
            root.right = newLeftNode
            return deleted,root
        
        if root.left ==None and root.right ==None:
            return True,None
        
        if root.left ==None:
            return True,root.right
        
        if root.right ==None:
            return True,root.left
        
        replacement = self.min(root.right)
        root.data = replacement
        deleted,newRightNode = self._delete(root.right,replacement)
        root.right = newRightNode
        return True,root
        
        
    def delete(self, data):
        deleted,newRoot = self._delete(self.root,data)
        if deleted:
            self.numNodes-=1
        self.root = newRoot
        return deleted
        

            
    
    def count(self):
        return self.numNodes 
```
