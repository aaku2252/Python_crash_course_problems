```python
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None




def deleteNode(head, pos):
    if head is None:
        return
    index = 0
    current = head
    while current.next and index < pos:
        previous = current
        current = current.next
        index += 1
    if index < pos:
        return
    elif index == 0:
        head = head.next
    else:
        previous.next = current.next
    return head



# To print the linked list.
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()

def takeInput() :
    head = None
    tail = None

    datas = [3,4,5,2,6,1,9,-1]

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1

    return head


head = deleteNode(takeInput(), 3)
printLinkedList(head)
```
